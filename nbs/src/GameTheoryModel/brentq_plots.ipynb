# Brent's algorithm tolerance

- `brentq(f, a, b, args=(), xtol, rtol, maxiter)` 
   - xtol=2e-12, 
   - rtol=8.881784197001252e-16 (cannot be smaller)

- satisfies `np.allclose(x, x0, atol=xtol, rtol=rtol)`
   - x = The exact root
   - x0 = the computed root

- `np.allclose(a, b, atol, rtol)`:
   - abs(a - b) <= (atol + rtol * absolute(b))
