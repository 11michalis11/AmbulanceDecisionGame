## Spanning Tree Perspective

```python
def build_graph_theoretic_model(lambda_a, lambda_o, mu, num_of_servers, threshold, system_capacity, parking_capacity, nodesize=2000, fontsize=12):

    Q = abg.markov.get_transition_matrix(lambda_a=lambda_a, lambda_o=lambda_o, mu=mu, num_of_servers=num_of_servers, threshold=threshold, system_capacity=system_capacity, parking_capacity=parking_capacity)
    all_states = abg.markov.build_states(threshold, system_capacity, parking_capacity)
    G = nx.DiGraph() 
    for i, origin_state in enumerate(all_states):
        for j, destination_state in enumerate(all_states):
            if Q[i,j] > 0:
                G.add_edge(origin_state, destination_state, color="blue", weight=Q[i,j])

    plt.figure(figsize=((system_capacity + 1) * 1.5, (parking_capacity + 1) * 1.5))
    pos = {state: [state[1], -state[0]] for state in all_states}
    nx.draw_networkx_nodes(
        G,
        pos,
        node_size=nodesize,
        nodelist=[state for state in all_states if state[1] < num_of_servers],
    )
    nx.draw_networkx_nodes(
        G,
        pos,
        node_size=nodesize,
        nodelist=[state for state in all_states if state[1] >= num_of_servers],
        node_color="red",
    )
    
    nx.draw_networkx_edges(G, pos, arrowstyle="fancy", arrowsize=60)
    nx.draw_networkx_labels(G, pos, font_size=fontsize)

    plt.axis("off")
    return G
```

```python
def number_of_spanning_trees_rooted_at_a_state(M, root=0):

    adjacency_matrix = np.zeros([len(M.nodes), len(M.nodes)])
    for row, u in enumerate(M.nodes):
        for col, v in enumerate(M.nodes):
            if (u,v) in M.edges:
                adjacency_matrix[row, col] += 1

    diagonal_out_degree_matrix = np.diag([j for i,j in M.out_degree])
    laplacian = diagonal_out_degree_matrix - adjacency_matrix
    
    return np.linalg.det(np.delete(np.delete(laplacian, root, 0), root, 1))
```

```python
num_of_servers = 1
threshold = 1
parking_capacity = 1
number_of_spanning_trees = []

for system_capacity in range(1,10):
    M = build_graph_theoretic_model(lambda_a=0.1, lambda_o=0.1, mu=0.2, num_of_servers=num_of_servers, threshold=threshold, system_capacity=system_capacity, parking_capacity=parking_capacity)
    number_of_spanning_trees.append(round(number_of_spanning_trees_rooted_at_a_state(M, root=0)))
    plt.show()
```

# Combinatorial Function for state probabilities

```python
def get_triangle_read_by_row(n, k):
    # T(n,k) = k * Bin(n, k+1) - Bin(n, k) + 1
    return k * math.comb(n, k+1) - math.comb(n, k) + 1


def get_product_matrix(n, k):
    # T(n,k) = Sum i=k...n,Binomial(i-1,k-1)*i
    return sum([math.comb(i-1, k-1) * i for i in range(k, n+1)])


def get_lambda_a_mu_values(system_capacity, num_of_servers=1, threshold=1, parking_capacity=1, just_coeff=False):
    if system_capacity > 1:
        lambda_a_mu_values = (sym_lambda_a + sym_mu) ** (system_capacity - 1)
    else:
        lambda_a_mu_values = 1
    return lambda_a_mu_values if not just_coeff else [math.comb(system_capacity - 1, term) for term in range(system_capacity)]


def get_lambda_a_o_values(system_capacity, num_of_servers=1, threshold=1, parking_capacity=1, just_coeff=False):
    if system_capacity > 2:
        lambda_a_o_values = sym.expand(sym_lambda_a * ((sym_lambda_a + sym_lambda_o) ** (system_capacity - 2) - sym_lambda_a ** (system_capacity - 2) - sym_lambda_o ** (system_capacity - 2)) + sym_lambda_a * sym_lambda_o ** (system_capacity - 2))
    else: 
        lambda_a_o_values = 0
    return lambda_a_o_values if not just_coeff else [math.comb(system_capacity - 2, term) for term in range(1, system_capacity - 1)]


def get_lambda_a1_o_mu_values(system_capacity, num_of_servers=1, threshold=1, parking_capacity=1, just_coeff=False):
    coefficients_lambda_a1_o_mu = [term for term in range(2,system_capacity - 1)]
    lambda_a1_o_mu_values = sum([coefficients_lambda_a1_o_mu[i] * sym_lambda_a * sym_lambda_o ** (system_capacity - 3 - i) * sym_mu ** (i+1) for i in range(system_capacity - 3)])
    return lambda_a1_o_mu_values if not just_coeff else coefficients_lambda_a1_o_mu


def get_lambda_a_o1_mu_values(system_capacity, num_of_servers=1, threshold=1, parking_capacity=1, just_coeff=False):
    coefficients_lambda_a_o1_mu = [
        math.comb(system_capacity-1, system_capacity-term-3) + 
        get_triangle_read_by_row(system_capacity-2, term + 1) 
        - 1 
        for term in range(system_capacity-4)
        ]
    lambda_a_o1_mu_values = sum([coefficients_lambda_a_o1_mu[i-1] * sym_lambda_a ** (system_capacity - 2 - i) * sym_lambda_o * sym_mu ** (i) for i in range(1, system_capacity - 3)])
    return lambda_a_o1_mu_values if not just_coeff else coefficients_lambda_a_o1_mu


def get_lambda_a_o_mu1_values(system_capacity, num_of_servers=1, threshold=1, parking_capacity=1, just_coeff=False):
    coefficients_lambda_a_o_mu1 = [
        math.comb(system_capacity-1, system_capacity-term-4) 
        - math.comb(system_capacity-3, term+1) 
        + get_product_matrix(system_capacity-4, term+2)
        for term in range(system_capacity-5)]
    lambda_a_o_mu1_values = sum([coefficients_lambda_a_o_mu1[i-2] * sym_lambda_a ** (system_capacity - 2 - i) * sym_lambda_o  ** i * sym_mu for i in range(2, system_capacity - 3)])
    return lambda_a_o_mu1_values if not just_coeff else coefficients_lambda_a_o_mu1


def get_P00_combinatorically(system_capacity, num_of_servers=1, threshold=1, parking_capacity=1):
    
    lambda_a_mu_values = get_lambda_a_mu_values(system_capacity, num_of_servers=1, threshold=1, parking_capacity=1)
    lambda_a_o_values = get_lambda_a_o_values(system_capacity, num_of_servers=1, threshold=1, parking_capacity=1)
    
    lambda_a1_o_mu_values = get_lambda_a1_o_mu_values(system_capacity, num_of_servers=1, threshold=1, parking_capacity=1)
    lambda_a_o1_mu_values = get_lambda_a_o1_mu_values(system_capacity, num_of_servers=1, threshold=1, parking_capacity=1)
    lambda_a_o_mu1_values = get_lambda_a_o_mu1_values(system_capacity, num_of_servers=1, threshold=1, parking_capacity=1)
    
    return lambda_a_mu_values + lambda_a_o_values + lambda_a1_o_mu_values + lambda_a_o1_mu_values + lambda_a_o_mu1_values
```

# Better Combinatorial Function

```python
def get_triangle_read_by_row(n, k):
    # T(n,k) = k * Bin(n, k+1) - Bin(n, k) + 1
    return k * math.comb(n, k+1) - math.comb(n, k) + 1


def get_product_matrix(n, k):
    # T(n,k) = Sum i=k...n,Binomial(i-1,k-1)*i
    return sum([math.comb(i-1, k-1) * i for i in range(k, n+1)])


def term_coefficient(p1, p2, p3):
    """This function breaks down the coefficients of each term in 6 cases. When the term includes the expression:
            - lambda_a^0: the term has a coefficient of 1
            - lambda_o^0: the term has a binomial coefficient of (C-1 CHOOSE p3)
            - mu^0: the term has a binomial coefficient of (C-2 CHOOSE p2)
        
        Now given that none of the above expressions occurs then the term may include:
            - lambda_a^1: The term has a coefficient of (C - p2 - 1)
            - lambda_o^1: The term has a coefficient of (C-1 CHOOSE p1) - 1 + Trinagle_sequence(C-2, p3)
            - mu^1: The term has a coefficient of (C-1 CHOOSE p1) - (C-3 CHOOSE (p2-1)) + Product_matrix(C-4, p2)
    """
    if p1 == 0 and p2 > 0:
        return 0
    if p1 == 0 and p2 == 0 and p3 > 0:
        return 1
    if p2 == 0 and p1 > 0:
        return math.comb(p1 + p3, p3)
    if p3 == 0 and p1 > 0 and p2 > 0:
        return math.comb(p1 + p2 - 1, p2)
    if p1 == 1:
        return p3 + 1
    if p2 == 1 and p1 > 1:
        return math.comb(p1 + p3 + 1, p1) - 1 + get_triangle_read_by_row(p1 + p3, p3) 
    if p3 == 1:
        return math.comb(p1 + p2 + 1, p1) - math.comb(p1 + p2 -1, p2 - 1) + get_product_matrix(p1 + p2 - 2, p2) 

    return sym.symbols("U" + str(p1) + str(p2) + str(p3))
```

```python
def get_00_rate_combinatorically(system_capacity, num_of_servers=1, threshold=1, parking_capacity=1):
    
    rate_00 = 0
    for p1 in np.linspace(system_capacity - 1, 0, system_capacity, dtype=int):
        for p2 in np.linspace(system_capacity - p1 - 1, 0, system_capacity - p1, dtype=int):
            p3 = system_capacity - p1 - p2 - 1
            
            rate_00 += term_coefficient(p1, p2, p3) * sym_lambda_a ** p1 * sym_lambda_o ** p2 * sym_mu ** p3

    return rate_00

get_00_rate_combinatorically(7)
```
