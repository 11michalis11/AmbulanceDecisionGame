```python
import ambulance_game as abg
import ciw
import numpy as np
import matplotlib.pyplot as plt
```

## PROBLEMS
- when threshold > system_capacity: states don't equate
- when threshold > num_of_servers: probabilities don't equate

```python
# Parameters
lambda_a = 0.2
lambda_o = 0.1
mu = 0.1
num_of_servers = 5
threshold = num_of_servers
system_capacity = 9
parking_capacity = 7
```

```python
# Parameters
seed_num = None     
runtime = 20000
warm_up_time = 10
num_of_trials = 1
output_type = "tuple"
tracker = ciw.trackers.NodePopulation()
```

```python
Q = abg.simulation.simulate_model(lambda_a, lambda_o, mu, num_of_servers, threshold, seed_num, runtime, system_capacity, parking_capacity, tracker=tracker)

transition_matrix = abg.markov.get_transition_matrix(lambda_a, lambda_o, mu, num_of_servers, threshold, system_capacity, parking_capacity)
pi = abg.markov.get_steady_state_algebraically(transition_matrix, algebraic_function=np.linalg.lstsq)
```

```python
all_states = abg.markov.build_states(threshold, system_capacity, parking_capacity)
```

```python
markov_states_probabilities_dictionary = {}
for i in range(len(all_states)):
    markov_states_probabilities_dictionary[all_states[i]] = pi[i]
```

```python
plt.figure(figsize=(20,10))
state_probabilities_array_sim = np.full((parking_capacity + 1, system_capacity + 1), np.NaN) 
for key, value in Q.statetracker.state_probabilities().items(): 
    if value > 0:
        state_probabilities_array_sim[key] = value 
plt.imshow(state_probabilities_array_sim)
plt.colorbar()

plt.figure(figsize=(20,10))
state_probabilities_array_markov = np.full((parking_capacity + 1, system_capacity + 1), np.NaN) 
for key, value in markov_states_probabilities_dictionary.items(): 
    if value > 0:
        state_probabilities_array_markov[key] = value 
plt.imshow(state_probabilities_array_markov)
plt.colorbar()
```

```python
plt.figure(figsize=(20,10))
plt.imshow(state_probabilities_array_markov - state_probabilities_array_sim)
plt.colorbar()
```

```python
def get_heatmap(state_probabilities_dictionary):
    plt.figure(figsize=(20,10))
    state_probabilities_array = np.full((parking_capacity + 1, system_capacity + 1), np.NaN) 
    for key, value in state_probabilities_dictionary.items(): 
        if value > 0:
            state_probabilities_array[key] = value 
    plt.imshow(state_probabilities_array)
    plt.colorbar()
    return state_probabilities_array
```

```python
# Parameters
lambda_a = 0.2
lambda_o = 0.1
mu = 0.1
num_of_servers = 5
threshold = num_of_servers
system_capacity = 9
parking_capacity = 7

Q = abg.simulation.simulate_model(lambda_a, lambda_o, mu, num_of_servers, threshold, seed_num, runtime, system_capacity, parking_capacity, tracker=tracker)

transition_matrix = abg.markov.get_transition_matrix(lambda_a, lambda_o, mu, num_of_servers, threshold, system_capacity, parking_capacity)
pi = abg.markov.get_steady_state_algebraically(transition_matrix, algebraic_function=np.linalg.lstsq)
all_states = abg.markov.build_states(threshold, system_capacity, parking_capacity)
```

```python
sim_states_array = get_heatmap(Q.statetracker.state_probabilities())
```

```python
markov_states_probabilities_dictionary = {}
for i in range(len(all_states)):
    markov_states_probabilities_dictionary[all_states[i]] = pi[i]
```

```python
markov_states_array = get_heatmap(markov_states_probabilities_dictionary)
```

```python
diff_states_array = sim_states_array - markov_states_array
plt.figure(figsize=(20,10))
plt.imshow(diff_states_array)
plt.colorbar()
```
